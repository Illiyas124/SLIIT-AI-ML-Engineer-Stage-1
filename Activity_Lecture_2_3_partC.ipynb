```python
import matplotlib.pyplot as plt
import numpy as np

# Define a custom function
def my_function(x):
    return np.sin(x) * np.exp(-0.1 * x)

# Generate x values
x = np.linspace(0, 20, 500)
y = my_function(x)

# Plot the function
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$f(x) = \sin(x) \cdot e^{-0.1x}$', color='blue')
plt.title("Custom Function Plot", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
